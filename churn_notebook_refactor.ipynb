```python
import os
import logging
import numpy as np
import matplotlib.pyplot as plt
import churn_library as cls
from churn_library import KEEPS_COLUMNS
```

```python
dataframe = cls.import_data(r"./data/bank_data.csv")
dataframe.head()
```

```python
cls.perform_eda(dataframe)
```

```python
CATEGORY_COLUMNS = [
    "Gender",
    "Education_Level",
    "Marital_Status",
    "Income_Category",
    "Card_Category",
]
encoded_dataframe = cls.encoder_helper(dataframe, CATEGORY_COLUMNS, "Churn")
```

```python
x_train, x_test, y_train, y_test = cls.perform_feature_engineering(encoded_dataframe, "Churn")
```

```python
best_model = cls.train_models(x_train, x_test, y_train, y_test)
```

```python
# Calculate feature importances
importances = best_model.feature_importances_
# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [dataframe[KEEPS_COLUMNS].columns[i] for i in indices]

# Create plot
plt.figure(figsize=(20,5))

# Create plot title
plt.title("Feature Importance")
plt.ylabel('Importance')

# Add bars
plt.bar(range(dataframe[KEEPS_COLUMNS].shape[1]), importances[indices])

# Add feature names as x-axis labels
plt.xticks(range(dataframe[KEEPS_COLUMNS].shape[1]), names, rotation=90);
```
